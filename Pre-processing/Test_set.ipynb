```python
# For running inference on the TF-Hub module.
import pandas as pd
import numpy as np

import tensorflow as tf

# For downloading the image.
import matplotlib.pyplot as plt

import cv2

# For drawing onto the image.
from PIL import Image

# For measuring the inference time.
import time
import os

from camtrapml.detection.models.megadetector import MegaDetectorV4_1
from camtrapml.detection.utils import render_detections
from camtrapml.image.utils import load_image, thumbnail
from camtrapml.detection.utils import extract_detections_from_image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
test_path = '/content/drive/MyDrive/ieee-final-round/test_features/test_features'
```

```python
test_names = sorted(os.listdir(test_path)) # the name are sorted
```

```python
len(test_names)
```

```python
test_clean_dir = '/content/drive/MyDrive/ieee-final-round/test_cleaned'
```

```python
def test_pipline() :
  for idx , name in enumerate(test_names):
    idx = idx + 665
    print(f"\r Current Idx {idx} " ,end='')
    path = os.path.join(test_path ,name )
    
    with MegaDetectorV4_1() as detector:
      detections = detector.detect(path)
    
    image = list(extract_detections_from_image(load_image(path), detections)) 
    if (len(image)==0): # there is detection 
      image = Image.open(path) 
    else : 
      image = image[0] 
    
    image = np.array(image)
    
    file_name = os.path.join(test_clean_dir,f'{name}') # the name had .png in it 
    cv2.imwrite(file_name,image)
    
```

```python
test_pipline()
```
